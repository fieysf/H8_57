## Area Pots (Series/Dataframe) <a id="12"></a>

**What is a line plot and why use it?**

An Area chart or area plot is a type of plot which displays information as a series of data points called 'markers' connected by straight line segments. It is a basic type of chart common in many fields. Use line plot when you have a continuous data set. These are best suited for trend-based visualizations of data over a period of time.

**Questions:**

Bagaimana grafik kriminalitas yang terjadi di London pada tahun 2008 sampai dengan 2016? dan apa yang menyebabkan turun dan naiknya kriminalitas?

```python
# First of all, Drop the Dataframe that we don't need, in this case i want to use year as an index so i drop everything except 'year' and 'value'
crime_drop_non_year = df_crmlnd.drop(['lsoa_code','borough','minor_category','month','major_category'], axis = 1)
crime_year = crime_drop_non_year.set_index('year') # Set 'year' as index
year = crime_year.groupby('year').sum() # Groupby the year and sum the value by the year 
# Graded-Funtion Begin (~1 Lines)
year.plot(kind='line', figsize=(14,8))
# Graded-Funtion End
plt.title('London Crime Data') # add a title to the area plot
plt.ylabel('Number of crime happened') # add y-label
plt.xlabel('Year') # add x-label

plt.show()
```

**Insight:**

From the graph, it can be seen that crime occurred in London from 2008 to 2016 has up and down trend line. from 2008 to 2010, the crime rate decreased and suddenly it got peaked in 2012. After 2012, Crime trend lowered until 2014 and in 2015 the crime trend increased and continued peaked in 2016.

## Histogram

A histogram is a way of representing the frequency distribution of numeric dataset. The way it works is it partitions the x-axis into bins, assigns each data point in our dataset to a bin, and then counts the number of data points that have been assigned to each bin. So the y-axis is the frequency or the number of data points in each bin. Note that we can change the bin size and usually one needs to tweak it so that the distribution is displayed nicely.

**Question:**

During 2016, where did the highest number of crimes occur?

```python
# The number of borough in 2016
borough_2016 = df_crmlnd[(df_crmlnd.year>=2016)]
totcrimes_borough_2016 = borough_2016.groupby('borough')['value'].sum()

print(totcrimes_borough_2016)

totcrimes_borough_2016.plot(kind="hist", figsize=(18, 6))
plt.title('Histogram of London Crime Data 2016') # add a title to the histogram
plt.ylabel('Num of Borough ') # add y-label
plt.xlabel('Num of Crimes') # add x-label
plt.show()
```

**Insight:**

During 2016, the highest number of crimes occurred in Westminster, Lambeth got the second position, and Southwark got the third.

## Bar Charts (Dataframe) <a id="10"></a>

A bar plot is a way of representing data where the *length* of the bars represents the magnitude/size of the feature/variable. Bar graphs usually represent numerical and categorical variables grouped in intervals. 

To create a bar plot, we can pass one of two arguments via `kind` parameter in `plot()`:

* `kind=bar` creates a *vertical* bar plot
* `kind=barh` creates a *horizontal* bar plot

**Question:**

what crime occured the most in 2015?

```python
# Set the year to 2015
year_2015 = df_crmlnd[df_crmlnd['year']==2015]
year_2015_drop = year_2015.drop(['lsoa_code','borough','minor_category','month'], axis=1) # Drop the series we don't need
year_2015_change_index = year_2015_drop.set_index('major_category') # Change the index into 'major_category'
mjrsumsfinal = year_2015_change_index.groupby('major_category').sum() # Group the major category then sums it
mjrsumsfinal.sort_values(by='value', ascending=False, axis=0, inplace=True) # Sorting the data
trmj = mjrsumsfinal.transpose() # Transposing the data
val = trmj.loc['value'] # Locating the 'value'
# Graded-Funtion Begin (~1 Lines)
val.plot(kind='barh', figsize=(12,6))
# Graded-Funtion End
plt.title('Crime Happened in 2015', y=1.1, size=14) # Add title to the plot
plt.xlabel('Number of Crime Happened', size=12) #Add x Label to the plot
plt.ylabel('Type of Crime', size=12) #Add y Label to the plot

plt.show()
```

**Insight:**

From the Bar Charts above, it is clear that Theft and Handling ranks first for the most frequent crimes in London in 2015. 

## Pie Charts <a id="6"></a>

A `pie chart` is a circualr graphic that displays numeric proportions by dividing a circle (or pie) into proportional slices. You are most likely already familiar with pie charts as it is widely used in business and media. We can create pie charts in Matplotlib by passing in the `kind=pie` keyword.

**Question:**

What crime occured the most for 8 years in London?

```python
# First of all, Drop the Dataframe that we don't need, in this case i want to use year as an index so i drop everything except 'major_category' and 'value'
crime_drop_non_mjr = df_crmlnd.drop(['lsoa_code','borough','minor_category','month','year'], axis = 1)
# Set 'major_category' as index
crime_drop_change_index = crime_drop_non_mjr.set_index('major_category')
# Groupby the 'Major Category' and sum the value by the 'Major Category'
major = crime_drop_change_index.groupby('major_category').sum()
# Plotting the Crime that categorized as Major Crime with piechart
#Selecting colors for the piechart
pie_colors = ["gold", "yellowgreen", "lightcoral", "lightskyblue", "lightgreen", "pink", "yellow", "limegreen", "silver"]
# ratio for each continent with which to offset each wedge.
explode_list = [.1, .1, .1, .1, .1, .1, .1, .1,.1]
major["value"].plot(kind="pie", 
                    figsize=(14, 8), 
                    shadow=True,
                    autopct="%1.1f%%", 
                    explode=explode_list, #'explode'
                    colors=pie_colors, 
                    pctdistance=1.12, # the ratio between the center of each pie slice and the start of the text generated by autopct
                    labels=None # turn off labels on pie chart
                   )
plt.legend(labels=major.index, loc="best")

# scale the title up by 12% to match pctdistance
plt.title('Crime in London by 2008 until 2016', y=1.12)

plt.axis('equal')

# add legend
plt.legend(labels=major.index, loc='lower right')

plt.show()
```

**Insight:**

Based on the Pie Chart above, Theft and Handling is the most common crime in London for 8 years. In my opinion, this happens the most because everyone from the young to the old age range can commit this crime. one example is stealing in a store. So that the high crime rate of theft and handling type for 8 years.

## Box Plots <a id="8"></a>

A `box plot` is a way of statistically representing the *distribution* of the data through five main dimensions: 

- **Minimun:** Smallest number in the dataset.
- **First quartile:** Middle number between the `minimum` and the `median`.
- **Second quartile (Median):** Middle number of the (sorted) dataset.
- **Third quartile:** Middle number between `median` and `maximum`.
- **Maximum:** Highest number in the dataset.

**Question:**

From 2010 to 2016, which year was the most crime at Greenwich ?

```python
# Write your function below
total_criminal_2011_2016 = crime[(crime.year>=2010) & (crime.year<=2016)] # Filtering the year from 2010 untul 2016
total_criminal_2011_2016_greenwich = total_criminal_2011_2016[(total_criminal_2011_2016.borough == 'Greenwich')] # Filtering the borough to Greenwich
num_of_crimes_greenwich_2010_2016 = total_criminal_2011_2016_greenwich.groupby('year')['value'].sum() # Grouping and sum the value by the year
greenwich = num_of_crimes_greenwich_2010_2016.T # Transposing the function
# Graded-Funtion Begin (~1 Lines)
greenwich.plot(kind="box", figsize=(8, 6))
# Graded-Funtion End

plt.title('Summary Statistic of Greenwich')
plt.ylabel('Number of Crimes')

plt.show()
```

**Insight:**

From 2010 to 2016, it turned out that in 2016 is the most crimes occurred in Greenwich.

## Scatter Plots <a id="10"></a>

A `scatter plot` (2D) is a useful method of comparing variables against each other. `Scatter` plots look similar to `line plots` in that they both map independent and dependent variables on a 2D graph. While the datapoints are connected together by a line in a line plot, they are not connected in a scatter plot. The data in a scatter plot is considered to express a trend. With further analysis using tools like regression, we can mathematically calculate this relationship and use it to predict trends outside the dataset.

**Question:**

Do the number of criminal cases depend on the month?

```python
# Write your function below

# Graded-Funtion Begin (~1 Lines)
df_crmlnd.plot(kind='scatter', x='month', y='value', figsize=(14,12)) # creating scatter plot

# Graded-Funtion End
plt.title("London's Monthly Criminal Cases in 2008-2016")
plt.xlabel('Month')
plt.ylabel('Number of Cases', labelpad=14)
plt.show()
```

## Word Clouds <a id="8"></a>


`Word` clouds (also known as text clouds or tag clouds) work in a simple way: the more a specific word appears in a source of textual data (such as a speech, blog post, or database), the bigger and bolder it appears in the word cloud.

```python
# import package and its set of stopwords
from wordcloud import WordCloud, STOPWORDS

print ('Wordcloud is installed and imported!')
```

```python
stopwords = set(STOPWORDS)
```

```python
# Picking 2000 random Samples of data 
crime_limit = crime.sample(n=2000, random_state = 1)
crime_limit.info # To be more clear about the sample
# Turning 'borough' into string
bor_word = crime_limit['borough'].to_string(index=False)
# instantiate a word cloud object
borough_wordcloud = WordCloud(background_color="white",
                            max_words=2000,
                            stopwords=STOPWORDS)
# generate the word cloud
borough_wordcloud.generate(bor_word)
```

```python
# Showing the wordcloud
plt.figure(figsize=(10, 5))
# Graded-Funtion Begin (~1 Lines)
plt.imshow(borough_wordcloud, interpolation="bilinear")
# Graded-Funtion End
plt.axis("off")
plt.show()
```
