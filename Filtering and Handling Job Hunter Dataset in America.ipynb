Kali ini saya akan memfilter dan handling sebuah dataset yang saya dapat dari Datascientist job Market di Amerika.

```python
import numpy as np
import pandas as pd
```

```python
list_job_object = ['position', 'company', 'description', 'numreview', 'location']
df[list_job_object]
```

```python
list_job_object = ['position', 'company', 'description', 'numreview', 'location']
df[list_job_object].describe()
```

```python
filter_null_all_columns = df.isnull().any(axis=1)
df[filter_null_all_columns]
```

```python
df_clean = df.copy()
df_clean['position'] = df['position'].fillna('blank',axis=0,inplace=False)
df_clean
```

```python
df_clean_company = df_clean.copy()
df_clean_company['company'] = df_clean['company'].fillna('blank',inplace=False)
df_clean_company
```

```python
df_clean_numreview = df_clean_company.copy()
df_clean_numreview['numreview'] = df_clean_company['numreview'].fillna('0 reviews',inplace=False)
df_clean_numreview
```

```python
df_clean_location = df_clean_numreview.copy()
df_clean_location['location'] = df_clean_numreview['location'].fillna('Blank',inplace=False)
df_clean_location
```
